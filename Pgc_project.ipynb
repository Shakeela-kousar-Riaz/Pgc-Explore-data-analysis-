```python
import pandas as pd
import plotly.express as px
```

```python
# 1. Bar chart comparing the total marks for different subjects
bar_data = pgc_data[['M_total_Marks', 'NM_total_marks', 'ICS_total_marks', 'FA_total_marks']].dropna()
fig1 = px.bar(bar_data, x=bar_data.index, y=['M_total_Marks', 'NM_total_marks', 'ICS_total_marks', 'FA_total_marks'],
              title='Total Marks Comparison', labels={'index': 'Entry', 'value': 'Total Marks', 'variable': 'Subject'})
fig1.update_layout(barmode='group')
```

```python
# 2. Box plot to visualize the distribution of percentage scores for each subject
fig2 = px.box(pgc_data, y=['M_Percentage', 'NM_percentage', 'ICS_percentage', 'FA_percentage'],
              title='Percentage Distribution by Subject', labels={'value': 'Percentage', 'variable': 'Subject'})

fig2.show()
```

```python
# 3. Pie chart showing the distribution of statuses for one of the subjects
pie_data = pgc_data['M_Status'].value_counts()
fig3 = px.pie(pie_data, values=pie_data.values, names=pie_data.index, title='M_Status Distribution')

fig3.show()
```

```python
# 4. Scatter plot to explore the relationship between total marks and percentages for a specific subject
scatter_data = pgc_data[['M_total_Marks', 'M_Percentage']].dropna()
fig4 = px.scatter(scatter_data, x='M_total_Marks', y='M_Percentage', title='M_total_Marks vs M_Percentage')
fig4.show()
```

```python

# Remove percentage symbol and convert to float
pgc_data['M_Percentage'] = pgc_data['M_Percentage'].str.rstrip('%').astype(float)

# 5. Histogram to visualize the distribution of percentage scores for a specific subject
hist_data = pgc_data['M_Percentage'].dropna()
fig5 = px.histogram(hist_data, x='M_Percentage', title='M_Percentage Distribution')

# Show the histogram plot
fig5.show()
```

```python
# Aggregate data for the sunburst plot
sunburst_data = pgc_data.groupby(['M_Status', 'NM_Status', 'ICS_status', 'FA_status']).size().reset_index(name='count')

# Define levels for the sunburst plot
levels = ['M_Status', 'NM_Status', 'ICS_status', 'FA_status']

# Create a sunburst plot
fig_sunburst = px.sunburst(sunburst_data, path=levels, values='count', title='Status Distribution')

# Show the sunburst plot
fig_sunburst.show()
```
